```python
#frames of video

import imageio
video_data = imageio.get_reader('/content/drive/MyDrive/New folder/video.mp4')

# Iterate through the frames and convert to NumPy arrays
Frames = []
for frame in video_data:
    Frames.append(frame)             # appending frame in frames

# Now 'frames' is a list of NumPy arrays representing each frame
print(Frames[0]) # Print the first frame as an example
```
